# torch.nn.functional 的应用

```python
import torch
import numpy as np
from torch.autograd import Variable
import torch.nn.functional as F
x_data=np.array([1.0,2.0,3.0,4.0])
y_data=np.array([0.0,0.0,1.0,1.0])
x_data=torch.Tensor(x_data.reshape(4,1))
y_data=torch.Tensor(y_data.reshape(4,1))
```

# 区别就是把以前线性变为sigmoid（线性）

# 激发器目前已不止sigmoid，ReLU,ReLU6,ELU,Tanh

```python
class Model(torch.nn.Module):
    def __init__(self):
        super(Model,self).__init__()
        self.linear=torch.nn.Linear(1,1)
    def forward(self,x):
        y_pred=F.sigmoid(self.linear(x))
        return y_pred

model=Model()
```

# logistic损失函数需要改crossentropy(多类别交叉熵)或者BCE（二分类交叉熵）

```python
criterion=torch.nn.BCELoss(reduction='sum')
optimizer=torch.optim.SGD(model.parameters(),lr=0.01)
```

```python
for epoch in range(4000):
    y_pred=model(x_data)
    loss=criterion(y_pred,y_data)
    print(epoch,loss.data[0])
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
```

```python
hour_var=Variable(torch.Tensor([[1.0]]))
print("predict 1 hour","1.0",model(hour_var).data[0][0]>0.5)
hour_var=Variable(torch.Tensor([[7.0]]))
print("predict 7 hour","7.0",model(hour_var).data[0][0]>0.5)
```
