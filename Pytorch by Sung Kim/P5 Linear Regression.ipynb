### x,y都是二维m×n张量，因此该问题为x，y需要变成3×1张量，而不是三个元素的一维张量

```python
import torch
import numpy as np
from torch.autograd import Variable
x_data=np.array([1.0,2.0,3.0])
y_data=np.array([2.0,4.0,6.0])
x_data=torch.Tensor(x_data.reshape(3,1))
y_data=torch.Tensor(y_data.reshape(3,1))
```

# STEP 1:Model Class in Pytorch

### 1)各种模型nn.Linear；nn.conv2d;nn.maxpool2d
### 2)相当于定义了model class里面的函数，包括model自己和model.forward(x)

```python
class Model(torch.nn.Module):
    def __init__(self):
        super(Model,self).__init__()
        self.linear=torch.nn.Linear(1,1)
    def forward(self,x):
        y_pred=self.linear(x)
        return y_pred

model=Model()
```

# STEP 2:Construct loss and optimizer

### 1)损失函数有多种，包括nn.L1loss,nn.SmoothL1Loss,nn.MSELoss


### 2)最优化方法有多种，包括optim.SGD,  optim.Adagrad, optim.Adam, optim.Adamax, optim.crossEntropyLoss

```python
criterion=torch.nn.MSELoss(reduction='sum')
optimizer=torch.optim.SGD(model.parameters(),lr=0.01)
```

# STEP 3:Training:forward,loss,backward,step

### 步骤：
###            1）先计算y预测值

###            2）先计算损失函数

###            3）梯度清理

###            4）反向传播计算梯度

###            5）参数优化

```python
for epoch in range(500):
    y_pred=model(x_data)
    loss=criterion(y_pred,y_data)
    print(epoch,loss.data[0])
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
```

# STEP 4:Test model

### 需要1*1的二维向量

```python
hour_var=Variable(torch.Tensor([[4.0]]))
print("predict(after training)","4",model.forward(hour_var).data[0][0])
```
